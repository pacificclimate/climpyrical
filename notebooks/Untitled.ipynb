```python
PATH = "/home/nannau/msc_spot_check.txt"
```

```python
data = pd.read_csv(PATH, sep=":", header=None)
```

```python
df = pd.DataFrame()

df['Climate_ID'] = data[1].str.slice(0,6)
df['Year'] = data[1].str.slice(7,10)
df['Month'] = data[1].str.slice(11,13)
df['Day'] = data[1].str.slice(14,16)
df['Element_num'] = data[1].str.slice(13, 16)
df = pd.concat([df]*24, ignore_index=True)
datum = pd.DataFrame()
start = 16
for i in range(24):
    end = start+7
    datum[i] =  data[1].str.slice(start, end)
    start+=7
```

```python
df['datum'] = datum.values.reshape(df.shape[0])
df
```
